```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

file_path = 'https://www.astrouw.edu.pl/ogle/ogle4/OCVS/lmc/cep/phot/I/OGLE-LMC-CEP-0001.dat '


data = pd.read_csv(file_path, delim_whitespace=True, header=None, names=['t', 'y', 'sigma_y'])


print(data)
```

```python
t = data['t'].values
y = data['y'].values

plt.plot(t,y)

fft = np.fft.fft(y)

magnitud = np.abs(fft)

fp=max(magnitud)
n=np.where(magnitud==fp)


print("Frecuencia principal en:",fp)

muestreo = 1 / (t[1] - t[0])
print("Frecuencia de muestreo:",muestreo)
```

```python
phi = np.mod(t*fp,1)

plt.figure(figsize=(10, 6))
plt.scatter(phi, y)
plt.xlabel('Fase')
plt.ylabel('Magnitud')
plt.title('Diagrama de fase')
plt.show()
```
